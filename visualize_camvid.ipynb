```python
import os
import random
import glob
import numpy as np

from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
camvid_train_image_dir = './SegNet/CamVid/train'
camvid_train_label_dir = './SegNet/CamVid/trainannot'

camvid_train_image_paths = glob.glob(os.path.join(camvid_train_image_dir, '*.png'))
camvid_train_label_paths = glob.glob(os.path.join(camvid_train_label_dir, '*.png'))

idx = random.randint(0, len(camvid_train_image_paths))

camvid_train_image_path = camvid_train_image_paths[idx]
camvid_train_label_path = camvid_train_label_paths[idx]

camvid_train_image = np.array(Image.open(camvid_train_image_path))
camvid_train_label = np.array(Image.open(camvid_train_label_path))

print('Train image shape', camvid_train_image.shape)
print('Label image shape', camvid_train_label.shape)
print(np.unique(camvid_train_label))
plt.imshow(camvid_train_image)
```
